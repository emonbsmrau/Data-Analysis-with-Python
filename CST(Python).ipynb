```python
def get_water_bill(num_gallons):
  if num_gallons>=30001:
    cost=10*(num_gallons/1000)
  elif num_gallons>=22001 and num_gallons<=30000:
    cost=7*(num_gallons/1000)
  elif num_gallons>=8001 and num_gallons<=22000:
    cost=6*(num_gallons/1000)
  elif num_gallons>0 and num_gallons<=8000:
    cost=5*(num_gallons/1000)
  return cost
```

```python
print(get_water_bill(10000))
```

```python
print(get_water_bill(25000))
```

```python
def percentage_growth(num_users,yrs_ago):
    growth=((num_users[-1]-num_users[-1-yrs_ago])/num_users[-1-yrs_ago]*100)
    return growth
```
